```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def get_suburls_per_main(main_pages):
    df = pd.DataFrame(columns=['Index', 'URL', 'Title', 'Link get request'])
    
    for main_page_url in main_pages:
        # Make a request to the main page URL
        response = requests.get(main_page_url)

        # Check if the request was successful
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            h2_tags = soup.find_all('h2')

            # Initialize the counter
            counter = 0

            for h2 in h2_tags:
                a_tags = h2.find_all('a')

                # Initialize a list for each iteration
                sub_urls = []
                titles = []

                for a in a_tags:
                    # Store the URL in the list
                    sub_urls.append(a['href'])

                    # Find the title associated with the current href
                    titles.append(a['title'])

                    # Add a 2-second delay between requests to be more respectful
                    time.sleep(2)

                # Increment the counter
                counter += 1

                # Append the data to the DataFrame
                df = pd.concat([df, pd.DataFrame({
                    'Index': [counter] * len(sub_urls),
                    'URL': sub_urls,
                    'Title': titles,
                    'Link get request': [main_page_url] * len(sub_urls)
                })])

        else:
            print(f"Failed to retrieve the page. Status code: {response.status_code}")

    return df.reset_index(drop=True)

# Example usage
waymo_pages = [
    'https://waymo.com/blog/search/label/company%20news',
    'https://waymo.com/blog/search/label/company%20news?updated-max=2020-05-12T09:00:00-07:00&max-results=20&start=20&by-date=false',
    'https://waymo.com/blog/search/label/company%20news?updated-max=2018-11-16T09:00:00-08:00&max-results=20&start=40&by-date=false'
]

result_df = get_suburls_per_main(waymo_pages)
print(result_df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium import webdriver

def get_suburl_per_main(main_page_url):
    df = pd.DataFrame(columns=['Index', 'URL', 'Title', 'Link get request'])

    # Use Selenium to interact with the webpage
    driver = webdriver.Chrome()  # You need to have the ChromeDriver executable in your PATH
    driver.get(main_page_url)
    time.sleep(1)
    
    # Initialize the previous height
    previous_height = driver.execute_script('return document.body.scrollHeight')

    while True:
        # Scroll to the bottom of the page
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        
        # Wait for some time for the content to load
        time.sleep(1)
        
        # Get the new height after scrolling
        new_height = driver.execute_script('return document.body.scrollHeight')
        
        # Break the loop if the page height doesn't change (i.e., no more content to load)
        if new_height == previous_height:
            break
        
        # Update the previous height
        previous_height = new_height

    # Get the page source after loading more articles
    page_source = driver.page_source
    driver.quit()

    # Continue with BeautifulSoup as before
    soup = BeautifulSoup(page_source, 'html.parser')
    div_tags = soup.find_all('div', class_='views-row')

    # Initialize the counter
    counter = 0
    sub_urls = []
    titles = []

    for div in div_tags:
        sub_url = div.find('a')['href']
        title = div.find('h3', class_='card__title').text.strip()
        sub_urls.append(sub_url)
        titles.append(title)

        time.sleep(2)

        # Increment the counter
        counter += 1

    # Append the data to the DataFrame
    df = pd.concat([df, pd.DataFrame({
        'Index': [i for i in range(1, counter + 1)],
        'URL': sub_urls,
        'Title': titles,
        'Link get request': [main_page_url] * counter
    })])

    return df.reset_index(drop=True)

# Example usage
main_page_url = 'https://motional.com/news'
result_df = get_suburl_per_main(main_page_url)
print(result_df)
```
