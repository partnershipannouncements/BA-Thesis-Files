```python
import pandas as pd
import spacy

# Load the English NLP model and create a Lemmatizer
nlp = spacy.load('en_core_web_md')
```

##### Lemmatization

```python
b = csv['Title']
csv['Title_Lemmatized'] = ''

# Create a list to store lemmatized titles
lemmatized_titles = []

for a in b:
    doc = nlp(a)
    doc_lemmatized = [token.lemma_ for token in doc]
    sentence_lemmatized = ' '.join(doc_lemmatized)
    lemmatized_titles.append(sentence_lemmatized)

# Assign the lemmatized titles list to the 'Title_Stemmed' column
csv['Title_Lemmatized'] = lemmatized_titles
```

#### Keywords

```python
import re

table = """
partner		
partnership
partnering
partnered
joint venture	
joint initiative	
alliance		
collaboration 	
collaborate
collaborated	
cooperation
cooperate
team with	
teaming with
teamed with	
team up
link up
license	
licensed
licensing	
work together	
worked together
working together	
work with	
worked with
working with	
join forces	
joined forces	
joint agreement
"""

# Process the table and create a list of strings
lexicon = []
for line in table.split('\n'):
    # Split the line into words
    words = re.findall(r'\S+', line)
    
    # Filter out empty words
    non_empty_words = [word for word in words if word]

    # Append words to the list if not separated by more than one whitespace
    if non_empty_words:
        lexicon.append(' '.join(non_empty_words))

# Print the list of strings

# List of elements
lexicon_unigram = []
lexicon_bigram = []

for word in lexicon:
    if ' ' in word:
        lexicon_bigram.append(word)
    else:
        lexicon_unigram.append(word)
```

```python
import pandas as pd
import spacy

# Assuming you already have a DataFrame called 'csv' with a column 'Title_Lemmatized_Lowercased'
# If not, replace 'csv' with the actual DataFrame name and column accordingly

# Load the English NLP model from SpaCy
nlp = spacy.load('en_core_web_md')

# List of keywords
keywords = ['partner', 'partnership', 'partnering', 'partnered', 'joint venture', 'joint initiative', 'alliance',
            'collaboration', 'collaborate', 'collaborated', 'cooperation', 'cooperate', 'team with', 'teaming with',
            'teamed with', 'team up', 'link up', 'license', 'licensed', 'licensing',
            'work together', 'worked together', 'working together', 'work with', 'worked with', 'working with',
            'join forces', 'joined forces', 'joint agreement']

# Create a new column to store matches
csv['Keyword_Matches'] = ''

# Iterate over rows in the 'Title_Lemmatized_Lowercased' column
for index, row in csv.iterrows():
    # Tokenize the headline using SpaCy
    tokens = nlp(row['Title_Lemmatized_Lowercased'])

    # Check for keyword matches
    keyword_matches = []
    for phrase in keywords:
        # Check for exact match for single-word keywords
        if ' ' not in phrase:
            keyword_matches.extend([token.lemma_ for token in tokens if token.lemma_ == phrase])
        else:
            # Check for multi-word phrases
            phrase_tokens = nlp(phrase)
            if len(phrase_tokens) == 2:
                bigram_matches = [''.join(tokens[i:i + 2].lemma_) for i in range(len(tokens) - 1)
                                  if ''.join(tokens[i:i + 2].lemma_) == phrase]
                keyword_matches.extend(bigram_matches)

    # Update the 'Keyword_Matches' column with the matched keywords
    csv.at[index, 'Keyword_Matches'] = ', '.join(keyword_matches)

# Filter the DataFrame to show only rows with keyword matches
matching_headlines = csv[csv['Keyword_Matches'] != '']

# Display the DataFrame with matching headline
```
